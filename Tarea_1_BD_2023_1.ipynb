## Crear tablas (36 pts.)

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "facciones" (6 pts)

```python
%%sql
CREATE TABLE "facciones"
(
    id INTEGER GENERATED ALWAYS AS IDENTITY,
    PRIMARY KEY (id),
    nombre VARCHAR (45) NOT NULL,
    descripcion VARCHAR (45) NOT NULL
)
```

### Tabla "trabajos" (6 pts)

```python
%%sql
CREATE TABLE "trabajos"
(
    id INTEGER GENERATED ALWAYS AS IDENTITY,
    PRIMARY KEY (id),
    descripcion VARCHAR (45) NOT NULL,
    nombre VARCHAR (45) NOT NULL,
    sueldo INTEGER NOT NULL
)
```

### Tabla "personas" (6 pts)

```python
%%sql
CREATE TABLE "personas"
(
    id INTEGER GENERATED ALWAYS AS IDENTITY, 
    PRIMARY KEY (id),
    nombre VARCHAR (45) NOT NULL,
    apellidos VARCHAR (45) NOT NULL,
    fecha_nacimiento DATE NOT NULL,
    infectado BOOL NOT NULL,
    conyugue INT,
    FOREIGN KEY (conyugue) REFERENCES "personas" (ID)
)
```

### Tabla "cuentas_bancarias" (6 pts)

```python
%%sql
CREATE TABLE "cuentas_bancarias"
(
    id INT GENERATED ALWAYS AS IDENTITY,
    tipo_de_cuenta VARCHAR (45) NOT NULL,
    banco VARCHAR (45) NOT NULL,
    saldo INT NOT NULL,
    id_persona INT, 
    PRIMARY KEY (id),
    FOREIGN KEY (id_persona) REFERENCES "personas" (ID)
)
```

### Tabla "persona_pertenece_faccion" (6 pts)

```python
%%sql
CREATE TABLE "persona_pertenece_faccion"
(
    id_persona INT,
    id_faccion INT, 
    PRIMARY KEY (id_persona, id_faccion),
    FOREIGN KEY (id_persona) REFERENCES "personas",
    FOREIGN KEY (id_faccion) REFERENCES "facciones"
)
```

### Tabla "persona_tiene_trabajo" (6 pts)

```python
%%sql
CREATE TABLE "persona_tiene_trabajo"
(
    id_persona INT,
    id_trabajo INT,
    estado BOOL NOT NULL,
    ultima_vez_realizado TIMESTAMP NOT NULL,
    PRIMARY KEY (id_persona, id_trabajo),
    FOREIGN KEY (id_persona) REFERENCES "personas",
    FOREIGN KEY (id_trabajo) REFERENCES "trabajos"
)
```

**Solución:** El codigo se divide entre dos tablas y un INNER JOIN entre ellas, en donde la primera tabla se compone de la cantidad de veces que se realiza un trabajo en todas las facciones (la cantidad de medicos, cientificos, etc, para cada una de las facciones). Esta tabla es utilizada en un INNER JOIN con la tabla que contiene el número mayor de trabajadores para una o mas profesiones, de este INNER JOIN se seleccionan con WHERE justamente las cantidades que corresponden a los maximos en cada una de las facciones, resultando así el trabajo más realizado por cada facción. 

```python
%%sql

/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT
*
FROM(
SELECT tabla2."Facción", tabla2."Trabajo", tabla2.cantidad
FROM(
    SELECT "Facción", "Trabajo", COUNT(*) as cantidad
    FROM(
        SELECT
        t.nombre as "Trabajo", f.nombre as "Facción"
        FROM 
            trabajos as t INNER JOIN persona_tiene_trabajo as pt ON t.id = pt.id_trabajo 
            INNER JOIN persona_pertenece_faccion as pf ON pf.id_persona = pt.id_persona
            INNER JOIN facciones as f ON f.id = pf.id_faccion) as tabla
        GROUP BY "Facción",  "Trabajo"
        ORDER BY "Facción",  "Trabajo" DESC) as tabla2
    

    INNER JOIN 
    (
    SELECT "Facción", MAX(cantidad) AS max_cantidad
        FROM(
            SELECT "Facción", "Trabajo", COUNT(*) as cantidad
            FROM(
                SELECT
                t.nombre as "Trabajo", f.nombre as "Facción"
                FROM 
                    trabajos as t INNER JOIN persona_tiene_trabajo as pt ON t.id = pt.id_trabajo 
                    INNER JOIN persona_pertenece_faccion as pf ON pf.id_persona = pt.id_persona
                    INNER JOIN facciones as f ON f.id = pf.id_faccion) as tabla3
            GROUP BY "Facción",  "Trabajo") as tabla4
    GROUP BY "Facción") as tabla5 ON tabla5."Facción" = tabla2."Facción"
    

    WHERE
    tabla2.cantidad = tabla5.max_cantidad) as tabla_final
```

**Solución:** _Se ocupan las tablas de personas, persona_pertenece a faccion y la de cuentas bancarias las cuales se unen mediante inner join para buscar el banco usado por cada persona de la faccion en la que se encuentra, luego se usa un COUNT donde empieza a contar las cuentas bancarias de cada faccion para luego solo mostrar la faccion junto con el banco mas usado por las personas de aquella faccion._

```python
%%sql

/* Reemplace estas queries por las que resuelven las consultas pedidas. */

SELECT DISTINCT ON (f.id) 
    f.nombre AS "Facción", 
    cb.banco AS "Banco"
FROM 
    facciones f 
    INNER JOIN persona_pertenece_faccion ppf ON f.id = ppf.id_faccion 
    INNER JOIN personas p ON ppf.id_persona = p.id 
    INNER JOIN cuentas_bancarias cb ON p.id = cb.id_persona 
WHERE 
    (
        SELECT 
            COUNT(*) 
        FROM 
            cuentas_bancarias cb2 
            INNER JOIN personas p2 ON cb2.id_persona = p2.id 
            INNER JOIN persona_pertenece_faccion ppf2 ON p2.id = ppf2.id_persona 
        WHERE 
            ppf2.id_faccion = ppf.id_faccion 
            AND cb2.banco = cb.banco 
    ) >= ALL (
        SELECT 
            COUNT(*) 
        FROM 
            cuentas_bancarias cb3 
            INNER JOIN personas p3 ON cb3.id_persona = p3.id 
            INNER JOIN persona_pertenece_faccion ppf3 ON p3.id = ppf3.id_persona 
        WHERE 
            ppf3.id_faccion = ppf.id_faccion 
        GROUP BY 
            cb3.banco
    ) 
```

Ejemplo:
   
|ID | Nombre | Apellido |

**Solución:** Se usa una combinación de LEFT y INNER JOIN'S para seleccionar las Id, el nombre y el apellido de la intersecciones que se producen entre las personas y las facciones. 
De las cuales se puede identificar las personas traidoras contantando la cantidad de veces que aparece su ID, siendo el caso de los traidores que su cantidad sea superior a 2, se seleccionan y se muestran en la tabla.

```python
%%sql

/* Reemplace estas queries por las que resuelven las consultas pedidas. */
SELECT
    "ID", "Nombre", "Apellido" 
FROM
(
SELECT
    "Apellido", COUNT("ID") as cantidad,"Nombre", "ID"
FROM
(
SELECT
    p.id as "ID", p.nombre as "Nombre", substring(apellidos, 0, strpos(apellidos, ' ')) as "Apellido"
FROM 
    personas as p LEFT JOIN persona_pertenece_faccion as pf ON p.id = pf.id_persona INNER JOIN 
    facciones as f ON f.id = pf.id_faccion) as tabla
GROUP BY
    "Nombre", "Apellido", "ID"
) as tabla2
WHERE 
    cantidad >= 2
```
